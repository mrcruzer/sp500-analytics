```python
# Import librarys

import pandas as pd 
import yfinance as yf

```

```python
sp500_url = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
tickers = sp500_url['Symbol'].tolist()
```

```python
# Defino las fechas
start_date='2000-01-01'
end_date='2023-02-23'
```

```python
# Creo un Dataframe vacio
sp500_data = pd.DataFrame()
```

```python
sp500_url
```

```python
tickers
```

```python
for symbol in tickers:
    print(f"Descargando datos para {symbol}...")
    data = yf.download(symbol, start=start_date, end=end_date)
    if len(data) > 0:
        data['Symbol'] = symbol
        sp500_data = sp500_data.append(data)
```

```python
sp500_data = sp500_data.reset_index()
```
